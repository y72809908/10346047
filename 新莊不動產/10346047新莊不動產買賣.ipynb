```python
df_ml = pd.merge(df2,df_region,left_index=True,right_index=True)  # 把左右二個index相同的dataframe(df2h and df_region)做合併
df_ml.head()
```

```python
#missing data
total = df_ml.isnull().sum().sort_values(ascending=False)
percent = (df_ml.isnull().sum()/df_ml.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
df_ml = df_ml.dropna().reset_index(drop=True)
df_ml
```
